# split the data set into independent (X) and dependent (Y) data set

```python
X=df.iloc[:,2:31].values
Y=df.iloc[:,1].values
```

# split the data set in to 80% tranning and 20% testing

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.20,random_state=0)
```

# scale the data

```python
from sklearn.preprocessing import StandardScaler# as
sc = StandardScaler()
```

```python
X_train=sc.fit_transform(X_train)
X_test=sc.fit_transform(X_test)
```

# create a function for the models

```python
def models(X_train,Y_train):
    #logistic regression
    from sklearn.linear_model import LogisticRegression
    log=LogisticRegression(random_state=0)
    log.fit(X_train,Y_train)
    #Decision Tree
    from sklearn.tree import DecisionTreeClassifier
    tree=DecisionTreeClassifier(criterion='entropy', random_state=0)
    tree.fit(X_train,Y_train)
    #Random forest classifier
    from sklearn.ensemble import RandomForestClassifier
    forest=RandomForestClassifier(n_estimators=10,criterion='entropy',random_state=0)
    forest.fit(X_train,Y_train)
    #print the model accuracy on the training data
    print('[0] Logistic Regressionn Training Accuracy:',log.score(X_train,Y_train))
    print('[1] Decision Tree Classifer Training Accuracy:',tree.score(X_train,Y_train))
    print('[2] Random Forest Classifer Training Accuracy:',forest.score(X_train,Y_train))
    
    
    return log, tree, forest
```

# getting all models

```python
model=models(X_train,Y_train)
```

# test model accuracy on test data on confusion matrix

```python
from sklearn.metrics import confusion_matrix
for i in range(len(model)):
    print('Model ',i)
    cm = confusion_matrix(Y_test,model[0].predict(X_test))

    TP=cm[0][0]
    TN=cm[1][1]
    FN=cm[1][0]
    FP=cm[0][1]
    print(cm)
    print('Testing Accuracy = ',(TP+TN)/(TP+TN+FN+FP))
```

# Another way to get metrics of the models

```python
from sklearn.metrics import classification_report
from sklearn.metrics import accuracy_score

for i in range(len(model)):
    print("Model",i)
    print(classification_report(Y_test,model[i].predict(X_test)))
    print(accuracy_score(Y_test,model[i].predict(X_test)))
    print()
```
